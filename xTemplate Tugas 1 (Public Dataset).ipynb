### Pertanyaan 1: Berapa banyak jumlah data pemakai pembayaran untuk tiap-tiap (credit card,boleto,voucher ,dll) dan visualisasikan datanya


```python
import pandas as pd
#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)

# Define the file paths
#customers_file_path = "E-Commerce Public Dataset/customers_dataset.csv"
#geolocation_file_path = "E-Commerce Public Dataset/geolocation_dataset.csv"
order_items_file_path = "E-Commerce Public Dataset/order_items_dataset.csv"
order_payments_file_path = "E-Commerce Public Dataset/order_payments_dataset.csv"
#order_reviews_file_path = "E-Commerce Public Dataset/order_reviews_dataset.csv"
orders_file_path = "E-Commerce Public Dataset/orders_dataset.csv"
#product_category_translation_file_path = "E-Commerce Public Dataset/product_category_name_translation.csv"
products_file_path = "E-Commerce Public Dataset/products_dataset.csv"
sellers_file_path = "E-Commerce Public Dataset/sellers_dataset.csv"

# Read the CSV files into pandas DataFrames
#eolocation_data = pd.read_csv(geolocation_file_path)
order_items_data = pd.read_csv(order_items_file_path)
order_payments_data = pd.read_csv(order_payments_file_path)
#order_reviews_data = pd.read_csv(order_reviews_file_path)
orders_data = pd.read_csv(orders_file_path)
#product_category_translation_data = pd.read_csv(product_category_translation_file_path)
products_data = pd.read_csv(products_file_path)
sellers_data = pd.read_csv(sellers_file_path)


#visualisasi data yang ini -------------------------------------------------------
#order_items_data
#order_payments_data  #ini yang di pakai
#visualisasi data yang ini -------------------------------------------------------



import matplotlib.pyplot as plt

# Task 1: Count and visualize payment_type
payment_type_counts = order_payments_data['payment_type'].value_counts()

# Plotting the data
plt.figure(figsize=(8, 6))
payment_type_counts.plot(kind='bar', color='skyblue')
plt.title('Payment Type Distribution')
plt.xlabel('Payment Type')
plt.ylabel('Number of Users')
plt.xticks(rotation=0)

# Adding counts above the bars
for i, count in enumerate(payment_type_counts):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

plt.show()
```

### Pertanyaan 2:Berapa kali angsuran pembayaran terjadi juga buat visualisasi datanya


```python
# Task 2: Visualize payment_installments
payment_installments_counts = order_payments_data['payment_installments'].value_counts().sort_index()

# Plotting the data
plt.figure(figsize=(10, 6))
payment_installments_counts.plot(kind='bar', color='salmon')
plt.title('Payment Installments Distribution')
plt.xlabel('Number of Installments')
plt.ylabel('Number of Orders')

# Adding counts above the bars
for i, count in enumerate(payment_installments_counts):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

plt.show()
```

### Pertanyaan 3:Buat visualisasi untuk interval berpa banyak uang yang di keluarkan tiap interval 49/50 di mulai dari angka 1


```python
# Task 3: Calculate and visualize payment_value distribution
payment_value_intervals = pd.cut(order_payments_data['payment_value'], bins=[0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 10000])
payment_value_counts = payment_value_intervals.value_counts().sort_index()

# Plotting the data
plt.figure(figsize=(12, 6))
payment_value_counts.plot(kind='bar', color='lightgreen')
plt.title('Payment Value Distribution')
plt.xlabel('Payment Value Intervals')
plt.ylabel('Number of Orders')

# Adding counts above the bars
for i, count in enumerate(payment_value_counts):
    plt.text(i, count + 5, str(count), ha='center', va='bottom')

plt.show()
```

### Pertanyaan 4:Hitung dan buat visualisasi data untuk banyaknya terjadi pembelian dalam setiap quartal dimulai dari 2016 awal

```python
#task 4
import pandas as pd
import matplotlib.pyplot as plt

# Read the orders dataset
#orders_data = pd.read_csv("path_to_your_orders_dataset.csv")

# Convert timestamp columns to datetime format
orders_data['order_purchase_timestamp'] = pd.to_datetime(orders_data['order_purchase_timestamp'])
orders_data['order_delivered_customer_date'] = pd.to_datetime(orders_data['order_delivered_customer_date'])

# Calculate delivery intervals in days
orders_data['delivery_interval'] = (orders_data['order_delivered_customer_date'] - orders_data['order_purchase_timestamp']).dt.days

# Create intervals starting from the year 2016 with a 3-month interval
orders_data['purchase_year_month'] = orders_data['order_purchase_timestamp'].dt.to_period("Q")

# Plotting the data
plt.figure(figsize=(12, 6))
interval_counts = orders_data['purchase_year_month'].value_counts().sort_index()
interval_counts.plot(kind='bar', color='purple')
plt.title('Order Purchase Timestamps Distribution (3-Month Intervals)')
plt.xlabel('Order Purchase Timestamp (3-Month Intervals)')
plt.ylabel('Number of Orders')

# Adding counts above the bars
for i, count in enumerate(interval_counts):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')

plt.show()
```

### Pertanyaan 5:Tampilkan data untuk waktu pengiriman entitas pembelian(barang,dll) beserta jumlahnya mulai dari 0 -35 hari

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named orders_data
# orders_data = pd.read_csv("path_to_your_orders_dataset.csv")

# Convert timestamp columns to datetime format
orders_data['order_purchase_timestamp'] = pd.to_datetime(orders_data['order_purchase_timestamp'])
orders_data['order_delivered_customer_date'] = pd.to_datetime(orders_data['order_delivered_customer_date'])

# Calculate delivery intervals in days
orders_data['delivery_interval'] = (orders_data['order_delivered_customer_date'] - orders_data['order_purchase_timestamp']).dt.days

# Determine a suitable range for delivery intervals (up to 55 days)
max_interval = 70
delivery_intervals = list(range(1, max_interval + 2, 2))

# Create histogram
hist, bins = np.histogram(orders_data['delivery_interval'], bins=delivery_intervals + [float('inf')])

# Plotting the data
plt.figure(figsize=(14, 6))
plt.bar(range(len(hist)), hist, color='green', align='center')
plt.title('Delivery Time Intervals Distribution')
plt.xlabel('Delivery Time Intervals')
plt.ylabel('Number of Deliveries')

# Adding counts above the bars
for i, count in enumerate(hist):
    plt.text(i, count + 10, str(count), ha='center', va='bottom')
```
